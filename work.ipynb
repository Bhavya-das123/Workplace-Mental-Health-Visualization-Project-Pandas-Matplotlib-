```python
# Remove rows with missing important values
df = df.dropna(subset=['Age', 'Gender', 'treatment'])
```

```python
# Clean gender column
df['Gender'] = df['Gender'].str.strip().str.lower()
gender_map = {
    'male': 'Male', 'm': 'Male', 'man': 'Male', 'male-ish': 'Male', 'maile': 'Male', 
    'malr': 'Male', 'mail': 'Male',
    'female': 'Female', 'f': 'Female', 'woman': 'Female', 'female (cis)': 'Female', 
    'femake': 'Female', 'femail': 'Female', 'cis female': 'Female', 'cis-female/femme': 'Female',
    'trans-female': 'Trans', 'trans woman': 'Trans', 'trans man': 'Trans', 
    'non-binary': 'Nonbinary', 'genderqueer': 'Nonbinary', 'other': 'Other'}
df['Gender'] = df['Gender'].replace(gender_map)
df.loc[~df['Gender'].isin(['Male', 'Female', 'Trans', 'Nonbinary']), 'Gender'] = 'Other'
print(df['Gender'].value_counts())
```

Visual Analysis of Workplace Mental Health Factors using  Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Gender Distribution (Bar Chart)

```python
gender_counts = df['Gender'].value_counts()

plt.figure(figsize=(6,4))
plt.bar(gender_counts.index, gender_counts.values, color=['skyblue','lightcoral','violet'])
plt.title("Gender Distribution in Mental Health Survey")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()

```

Treatment Proportion (Pie Chart)

```python

treat_counts = df['treatment'].value_counts()

plt.figure(figsize=(5,5))
plt.pie(treat_counts.values, labels=treat_counts.index, autopct='%1.1f%%', startangle=140, colors=['b','g'])
plt.title("Overall Treatment Proportion")
plt.show()
```

Age Distribution (Histogram)

```python

plt.figure(figsize=(7,4))
plt.hist(df['Age'], bins=15, color='lightgreen', edgecolor='black')
plt.title("Age Distribution of Respondents")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Age vs Treatment (Boxplot)

```python

plt.figure(figsize=(7,4))
df.boxplot(column='Age', by='treatment', grid=False, patch_artist=True,
           boxprops=dict(facecolor='lightblue'))
plt.title("Age Distribution by Treatment Status")
plt.suptitle("")
plt.xlabel("Treatment")
plt.ylabel("Age")
plt.show()
```

Company Size vs Treatment (Bar Chart)

```python

company_treatment = df.groupby('no_employees')['treatment'].value_counts().unstack().fillna(0)

company_treatment.plot(kind='bar', figsize=(8,5), color=['black', 'yellow'])
plt.title("Treatment Count by Company Size")
plt.xlabel("Number of Employees")
plt.ylabel("Count")
plt.xticks(rotation=30)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.legend(title="Treatment")
plt.show()
```

Remote Work vs Treatment (Bar Chart)

```python

remote_treatment = df.groupby('remote_work')['treatment'].value_counts().unstack().fillna(0)

remote_treatment.plot(kind='bar', figsize=(6,4), color=['orchid','b'])
plt.title("Remote Work vs Treatment")
plt.xlabel("Remote Work")
plt.ylabel("Count")
plt.legend(title="Treatment")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

Work Interference vs Treatment (Bar Chart)

```python

interfere_treatment = df.groupby('work_interfere')['treatment'].value_counts().unstack().fillna(0)

interfere_treatment.plot(kind='bar', figsize=(8,5), color=["Pink",'Blue'])
plt.title("Work Interference vs Treatment")
plt.xlabel("Work Interference Level")
plt.ylabel("Count")
plt.legend(title="Treatment")
plt.xticks(rotation=30)
plt.show()
```

Combined View (Subplots)

```python

fig, axs = plt.subplots(2, 2, figsize=(12,10))

# 1️⃣ Gender Distribution
axs[0,0].bar(gender_counts.index, gender_counts.values, color='lightblue')
axs[0,0].set_title("Gender Distribution")
axs[0,0].set_xlabel("Gender")
axs[0,0].set_ylabel("Count")

# 2️⃣ Treatment Pie
axs[0,1].pie(treat_counts.values, labels=treat_counts.index, autopct='%1.1f%%', startangle=140, colors=["#b23131",'#66b3ff'])
axs[0,1].set_title("Treatment Proportion")

# 3️⃣ Age Histogram
axs[1,0].hist(df['Age'], bins=15, color='lightgreen', edgecolor='black')
axs[1,0].set_title("Age Distribution")
axs[1,0].set_xlabel("Age")
axs[1,0].set_ylabel("Count")

# 4️⃣ Remote Work vs Treatment
remote_treatment.plot(kind='bar', ax=axs[1,1], color=['#ffb3ba','#baffc9'])
axs[1,1].set_title("Remote Work vs Treatment")
axs[1,1].set_xlabel("Remote Work")
axs[1,1].set_ylabel("Count")

plt.tight_layout()
plt.show()
```

ANALYSIS

Most respondents belong to the 18–35 age range.

Males make up the majority, but females report higher treatment rates.

Employees in smaller companies and remote workers show a higher tendency to seek treatment.

Work interference is a major factor influencing mental health treatment needs.

Overall Objective

To explore how demographics, workplace environment, and organizational support influence the mental health and treatment decisions of employees in the tech and non-tech industries.
